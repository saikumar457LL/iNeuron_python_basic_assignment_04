17.How do you distinguish between copy.copy() and copy.deepcopy()?

Ans:

    copy.copy()  --> it copy's the data and creates another "refrence" to shallow data types of parent memory address. If any changes in shallow data types both parent and child will be effected.
    
    
    
    copy.deepcopy()  --> Every element in the deepcopy has it's own memory address and unique instance other than parent ones. It is independent, once it's copy the data from parent.
    If any thing is changed in parent one, it is not reflected in child, because child and parent has unique memory address. this rule is also applicable to child and parent.
